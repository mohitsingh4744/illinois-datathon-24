```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

cramer_v_score = cramers_v(combined_df['reason'], combined_df['same_day_recalls'])
cramer_v_score
```

```python
from scipy.stats import f_oneway
result = f_oneway(*(combined_df['recalls'][combined_df['reason'] == category] for category in combined_df['reason'].unique()))

print("ANOVA F-statistic:", result.statistic)
print("ANOVA p-value:", result.pvalue)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

labels = values[0:7]
labels.append("Other")

proportions = []
for i in range(7):
    proportions.append(counts[i]/len(all_recalls))

other_proportion = (len(all_recalls) - sum(counts[0:7]))/len(all_recalls)
proportions.append(other_proportion)

sorted_indices = sorted(range(len(proportions)), key=lambda i: proportions[i], reverse=True)
sorted_proportions = sorted(proportions, reverse=True)

sorted_labels = [labels[i] for i in sorted_indices]

explode = [0] * len(proportions)
explode[0] = 0.06 

colors = sns.color_palette('Spectral', n_colors = 8)  

plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(sorted_proportions, explode=explode, colors=colors, labels=None, autopct=lambda p: '{:.1f}%'.format(p), pctdistance=0.85, startangle=180)

for wedge in wedges:
    wedge.set_edgecolor('black')

plt.axis('equal')

for autotext in autotexts:
    autotext.set_horizontalalignment('center')
    autotext.set_fontweight('bold')

plt.title('Recalls by Call Reason', fontsize=14, fontweight='bold', pad=-10, color='black')

legend = plt.legend(wedges[::-1], sorted_labels[::-1], title="Call Reason", loc="center left", bbox_to_anchor=(0.9, 0, 0.5, 1.1))
legend.get_frame().set_linewidth(1)  
legend.get_frame().set_edgecolor('black')

plt.show()
```
